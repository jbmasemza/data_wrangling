### Load required libraries

```python
import pandas as  pd
import numpy as np
import re

```

### Load and read in personality Data Frame

```python
df = pd.read_csv("personality_scores.csv", delimiter = ';')
df.head()
```

### Exemine the data set

```python
print('Number of responsers are : ',len(df))
```

### Checking missing values

```python
null_values= df.isnull().sum().sum()
null_values
```

### Examining the data frame for duplicates (based on ID),

```python
df_duplicate = df[df.duplicated(['ID'])]
df_duplicate
```

### Dropping  duplicates and NAN values

```python
df.drop_duplicates(subset=['ID'], inplace=True)
df = df.dropna(axis= 1, how='all')
df.head()
```

###  Assert statement to check that the new data frame is the length of the unique entries of the original data frame.

```python
#Assertation
df['ID'].unique().sum() == df['ID'].sum()
```

### Total score of each of the personality test subscales

```python

category = dict({1:'Extraversion', 2:'Agreeableness', 3:'Conscientiousness', 4:'Emotional Stability', 5:'Intellect'})

def score(entry, category):
    numbers=0
    _r= re.compile('\d+')
    for col in entry:
        num= _r.findall(col)
        if str(category)==num[0]:
            numbers+= int(num[1])
        else:
            numbers= numbers
    return numbers
    
for i in category.keys():
    df[f'{category[i]}']= df.loc[:,'Section 5 of 6 [I am always prepared.]':'Section 5 of 6 [I worry about things.]'].apply(score, args=(i,) , axis=1)
    
df.head()
```

### Load and read in department Data Frame

```python
department = pd.read_csv("departments.csv", sep=";")
```

### Merging the personality score and the department data frame

```python
#merging department with personality scores
score_dep = pd.merge(df, department ,how='left')
score_dep.head()
```
